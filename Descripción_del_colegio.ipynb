```python
class Pared:
    def __init__(self, orientacion, color_fondo, decoraciones, tiene_ventana=False, tiene_puerta=False):
        self.orientacion = orientacion
        self.color_fondo = color_fondo
        self.decoraciones = decoraciones  # Lista de elementos decorativos
        self.tiene_ventana = tiene_ventana
        self.tiene_puerta = tiene_puerta

    def describir(self):
        print(f"--- Pared {self.orientacion} ---")
        print(f"Color de fondo: {self.color_fondo}")
        print(f"Decoraciones: {', '.join(self.decoraciones)}")
        if self.tiene_puerta: print(">> Tiene una puerta abierta.")
        if self.tiene_ventana: print(">> Tiene una ventana funcional.")
        print("")

class Colegio:
    def __init__(self):
        self.frontal = Pared("Frontal", "Amarillo", ["Arcoíris", "Nubes", "Reloj en la parte superior de la puerta", "Tiene un letrero 'Colegio'"], tiene_puerta=True)
        self.derecha = Pared("Derecha", "Piel/Salmón", ["Maceta con flores rojas"], tiene_ventana=True)
        self.izquierda = Pared("Izquierda", "Amarillo claro", ["Hay una mariposa de color rosa y azul", "Camino de puntos de la mariposa"], tiene_ventana=True)
        self.trasera = Pared("Trasera", "Anaranjado", ["Tiene 2 lápices gigantes azul y rojo", "Sol", "Hay pasto verde"], tiene_ventana=True)

        self.techo = "Marron con diseño de tejas"

    def mostrar_detalles(self):
        print(f"Detalles de: EL COLEGIO")
        print(f"Techo: {self.techo}\n")
        self.frontal.describir()
        self.derecha.describir()
        self.izquierda.describir()
        self.trasera.describir()

mi_colegio = Colegio()
mi_colegio.mostrar_detalles()
```
