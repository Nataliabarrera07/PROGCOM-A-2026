```python
def menu():
    Comida = ["combo 1","combo 2", "combo 3", "combo 4","papas","jugos naturales","cerveza","agua","gaseosas"]
    precios = [20000,25000,18000,24500,3000,7000,3500,4000,2500]
    for i in range(len(Comida)):
        print(Comida[i], "-", precios[i])
    return Comida, precios


total_dia = 0
factura = []

user = "s"

while user == "s":
    Comida, precios = menu()
    r = "s"

    while r == "s":
        x = input("¿Qué desea?: ")

        if x in Comida:
            cantidad = int(input(f"¿Cuántas unidades de {x} quieres pedir?: "))
            precio_unitario = precios[Comida.index(x)]
            subtotal = precio_unitario * cantidad

            total_dia += subtotal


            factura.append((x, cantidad, precio_unitario, subtotal))

        else:
            print("Producto no válido")

        r = input("¿Desea algo más? (s/n): ")

    user = input("¿Hay más usuarios? (s/n): ")


print("\n FACTURA ")
for item in factura:
    nombre, cantidad, precio, subtotal = item
    print(f"{nombre} x{cantidad} - ${precio} = ${subtotal}")

print("///////////")
print(f"TOTAL DEL DÍA: ${total_dia}")
```
