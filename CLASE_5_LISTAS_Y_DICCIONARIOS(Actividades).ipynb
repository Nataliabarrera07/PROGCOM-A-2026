````python
suma=0
c=0
notas=[]
n=int(input("Ingrese la cantidad de estudiantes"))
for i in range(n):
  nota=float(input(f"Ingrese la nota del estudiante {i+1}"))
  notas.append(nota)
  suma+=nota
prom=suma/n
for i in range(n):
  if notas[i]>prom:
    c=c+1
    print(f"El estudiante {i+1} tiene una nota mayor al promedio")
print(f"El promedio es {prom}")
````

````python

suma=0
c=0
salario=[]
n=int(input("Ingrese la cantidad de trabajadores"))
for i in range(n):
  s=float(input(f"Ingrese el sueldo del trabajador {i+1}"))
  salario.append(s)
  suma+=s
prom=suma/n
for i in range(n):
  if salario[i]>prom:
    c=c+1
    print(f"El trabajador {i} tiene un salario mayor al promedio")
print(f"El promedio es {prom}")
````

````python
##EJERCICIO 1. LISTA DE N TRABAJADORES
n = int(input("Ingrese la cantidad de trabajadores: "))

salarios = []

for i in range(n):
 sueldo = float(input(f"Ingrese el salario del trabajador {i+1}: "))
 salarios.append(sueldo)

promedio = sum(salarios) / n

print("El salario promedio es:", promedio)

mayores = []

for i in range(n):
 if salarios[i] > promedio:
  mayores.append(i+1)

print("Cantidad que ganan más que el promedio:", len(mayores))
print("Trabajadores que ganan más que el promedio:", mayores)
````

````python
##EJERCICIO 3
cantidad = int(input("Cuántos números tendrá la lista: "))

numeros = []

for i in range(cantidad):
 num = int(input("Ingrese un número: "))
 numeros.append(num)

n = int(input("Ingrese el número para verificar múltiplos: "))

contador = 0

for num in numeros:
 if num % n == 0:
  contador += 1

print("Cantidad de múltiplos de", n, ":", contador)
````
