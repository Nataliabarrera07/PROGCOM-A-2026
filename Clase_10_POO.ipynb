````python
class disfraz():
  #creando el constructor
  def __init__(self,color,talla,forma,material,personaje):
    #creando atributos de clase
    self.color=color
    self.talla=talla
    self.forma=forma
    self.material=material
    self.personaje=personaje

  def imprimir (self):
      return f"El disfraz es de {self.personaje}.\n Es de talla {self.talla}.\n Es de color {self.color}.\n Es de forma {self.forma}.\n Es de material {self.material}."
````

````python
class Edificio():
  #contruyendo el constructor
  def __init__(self,material,ubicacion,fachada,ventanas,puertas,nombre):
    self.material=material
    self.ubicacion=ubicacion
    self.fachada=fachada
    self.ventanas=ventanas
    self.puertas=puertas
    self.nombre=nombre

  def direccion(self,tipo):
    return f"Tipo de edificación:{tipo}\nDireccion:{self.ubicacion}"

  def descripcion(self,tipo):
    print(f"{self.nombre} es un {tipo} ubicado en {self.ubicacion}")
    print(f"Tiene {self.ventanas} ventana(s), {self.puertas} puerta(s)")
    print(f"Su fachada es {self.fachada}")
````

````python
colegio=Edificio("Ladrillo","Bucaramanga", "De colores pasteles con números","2","1","Los angelitos")
floristeria=Edificio("Ladrillo", "Floridablanca", "campestres con césped verde",0,1, "Florecitas")
````

````python
colegio.decripción("colegio")
````
