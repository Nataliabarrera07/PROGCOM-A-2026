````python
suma=0
c=0
notas=[]
n=int(input("Ingrese la cantidad de estudiantes"))
for i in range(n):
  nota=float(input(f"Ingrese la nota del estudiante {i+1}"))
  notas.append(nota)
  suma+=nota
prom=suma/n
for i in range(n):
  if notas[i]>prom:
    c+=1
print(f"El estudiante {i+1} tiene una nota mayor al promedio")
print(f"El promedio es {prom}")
````

````python
#votos de los 5 candidatos
votos = [0, 0, 0, 0, 0]

n = int(input("Cuantos usuarios votan: "))

# primera votación
for i in range(n):
voto = int(input("Vote por un candidato (1-5): "))

if voto >= 1 and voto <= 5:
votos[voto-1] += 1
else:
print("Voto invalido")

# mostrar votos
print("Resultados primera vuelta:")
for i in range(5):
print("Candidato", i+1, ":", votos[i], "votos")

# encontrar los dos mayores
primero = votos.index(max(votos))
votos_temp = votos.copy()
votos_temp[primero] = -1
segundo = votos_temp.index(max(votos_temp))

print("Pasan a segunda vuelta:")
print("Candidato", primero+1)
print("Candidato", segundo+1)

# segunda votación
votos2 = [0, 0]

print("Segunda vuelta")

for i in range(n):
voto = int(input(f"Vote por {primero+1} o {segundo+1}: "))

if voto == primero+1:
votos2[0] += 1
elif voto == segundo+1:
votos2[1] += 1
else:
print("Voto invalido")

# resultados finales
print("Resultados finales:")

print("Candidato", primero+1, ":", votos2[0], "votos")
print("Candidato", segundo+1, ":", votos2[1], "votos")

if votos2[0] > votos2[1]:
print("Ganador: Candidato", primero+1)
else:
print("Ganador: Candidato", segundo+1)
````
