```python
class Participante:
    def __init__(self, id_usuario, nom, ape):
        self.identificacion = id_usuario
        self.nombre = nom
        self.apellido = ape

    def actualizar_nombre(self, nuevo_nom):
        self.nombre = nuevo_nom

    def actualizar_apellido(self, nuevo_ape):
        self.apellido = nuevo_ape

    def presentar_datos(self):
        return f"ID: {self.identificacion} | {self.nombre} {self.apellido}"


class Estudiante(Participante):
    def __init__(self, id_usuario, nom, ape, programa, ciclo, tel, dir_residencia):
        super().__init__(id_usuario, nom, ape)

        self.carrera = programa
        self.semestre = ciclo
        self.telefono = tel
        self.direccion = dir_residencia

    def modificar_carrera(self, nueva_carrera):
        self.carrera = nueva_carrera

    def ficha_completa(self):
        info_basica = self.presentar_datos()
        return (f"{info_basica}\nCarrera: {self.carrera} | Semestre: {self.semestre}\n"
                f"Contacto: {self.telefono} | Dir: {self.direccion}")


class Asignatura:
    def __init__(self, titulo, tutor, cupo_max):
        self.nombre_materia = titulo
        self.profesor = tutor
        self.limite_estudiantes = cupo_max
        self.nomina_alumnos = []

    def inscribir_alumno(self, alumno_obj):
        if len(self.nomina_alumnos) < self.limite_estudiantes:
            self.nomina_alumnos.append(alumno_obj)
            print(f"✅ Éxito: {alumno_obj.nombre} ha sido inscrito en {self.nombre_materia}.")
        else:
            print(f"❌ Error: No hay cupos para {alumno_obj.nombre} en esta materia.")

    def registrar_calificacion(self, alumno_obj, valor_nota):
        print(f"📝 Calificación - {self.nombre_materia}:")
        print(f"Estudiante: {alumno_obj.nombre} {alumno_obj.apellido} -> Nota: {valor_nota}")

    def listar_grupo(self):
        print(f"\n--- LISTA DE CLASE: {self.nombre_materia} ---")
        for i, est in enumerate(self.nomina_alumnos, 1):
            print(f"{i}. {est.nombre} {est.apellido} ({est.carrera})")



if __name__ == "__main__":
    alumno1 = Estudiante(2431923611, "Harry", "Herrera", "Mecatronica", 4, "300123456", "Calle 10 #2-30")
    alumno2 = Estudiante(123476943, "Laura", "Diaz", "Economia", 2, "315987654", "calle 200 #30-12")

    alumno1.actualizar_nombre("Harry Andres")
    alumno2.modificar_carrera("Ingenieria Industrial")

    materia_python = Asignatura("Electromagnetismo", "Dr. Ramirez", 2)

    materia_python.inscribir_alumno(alumno1)
    materia_python.inscribir_alumno(alumno2)

    materia_python.listar_grupo()

    print("\n--- Reporte de Notas ---")
    materia_python.registrar_calificacion(alumno1, 5.0)
    materia_python.registrar_calificacion(alumno2, 5.0)
```
