```python
import math

a = float(input("Valor de a: "))
b = float(input("Valor de b: "))
c = float(input("Valor de c: "))

discriminante = b**2 - 4*a*c

if discriminante < 0:
    print("La ecuación no tiene soluciones reales.")
else:
    x1 = (-b + math.sqrt(discriminante)) / (2*a)
    x2 = (-b - math.sqrt(discriminante)) / (2*a)
    print(f"Soluciones: x1 = {x1}, x2 = {x2}")
```

```python
a = float(input("Lado A: "))
b = float(input("Lado B: "))
c = float(input("Lado C: "))

# Validación de existencia
if (a + b > c) and (a + c > b) and (b + c > a):
    if a == b == c:
        tipo = "Equilátero"
    elif a == b or a == c or b == c:
        tipo = "Isósceles"
    else:
        tipo = "Escaleno"
    print(f"Es un triángulo válido y es de tipo {tipo}.")
else:
    print("No se puede formar un triángulo con esas medidas.")
```

```python
import math

a = float(input("Valor de a: "))
b = float(input("Valor de b: "))
c = float(input("Valor de c: "))

disc = b**2 - 4*a*c

if disc > 0:
    x1 = (-b + math.sqrt(disc)) / (2*a)
    x2 = (-b - math.sqrt(disc)) / (2*a)
    print(f"Dos soluciones reales: {x1} y {x2}")
elif disc == 0:
    x = -b / (2*a)
    print(f"Una solución real (doble): {x}")
else:
    print("La ecuación tiene soluciones imaginarias.")
```
