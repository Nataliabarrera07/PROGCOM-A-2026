#Ejercicios del 16-25 usando WHILE

```python
n = int(input("Tabla del: "))
i = 1
while i <= 10:
    print(f"{n} x {i} = {n * i}")
    i += 1
```

```python
n = int(input("Suma hasta: "))
suma, i = 0, 1
while i <= n:
    suma += i
    i += 1
print(f"Suma: {suma}")
```

```python
n = int(input("Factorial de: "))
fact, i = 1, 1
while i <= n:
    fact *= i
    i += 1
print(f"Factorial: {fact}")
```

```python
n = int(input("Cant. Fibonacci: "))
a, b, c = 0, 1, 0
while c < n:
    print(a, end=" ")
    a, b = b, a + b
    c += 1
```

```python
secreto = "rojo"
intentos = 0
while intentos < 10:
    rta = input("Adivina el color: ")
    if rta == secreto:
        print("¡Ganaste!")
        break
    intentos += 1
```

```python
i = 1
while i <= 100:
    if i % 2 == 0:
        print(i, end=" ")
    i += 1
```

```python
n = int(input("\nNúmero para sumar dígitos: "))
suma = 0
while n > 0:
    suma += n % 10
    n //= 10
print(f"Suma dígitos: {suma}")
```

```python
n = int(input("¿Es primo?: "))
i, primo = 2, True
if n < 2: primo = False
while i < n:
    if n % i == 0:
        primo = False
        break
    i += 1
print(f"Resultado: {primo}")
```

```python
n = int(input("Filas triángulo: "))
i = 1
while i <= n:
    print("*" * i)
    i += 1
```

```python
b = int(input("Base: "))
e = int(input("Exponente: "))
res, i = 1, 0
while i < e:
    res *= b
    i += 1
print(f"Potencia: {res}")
```

#Ejercicios del 16-25 usando For

```python
n = int(input("Tabla del: "))
for i in range(1, 11):
    print(f"{n} x {i} = {n * i}")
```

```python
n = int(input("Suma hasta: "))
suma = 0
for i in range(1, n + 1):
    suma += i
print(f"Suma: {suma}")
```

```python
n = int(input("Factorial de: "))
fact = 1
for i in range(1, n + 1):
    fact *= i
print(f"Factorial: {fact}")
```

```python
n = int(input("Cant. Fibonacci: "))
a, b = 0, 1
for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
secreto = "morado"
for i in range(1, 11):
    rta = input(f"Intento {i}: ")
    if rta == secreto:
        print("¡Ganaste!")
        break
```

```python
for i in range(2, 101, 2):
    print(i, end=" ")
```

```python
n_str = input("\nNúmero para sumar dígitos: ")
suma = 0
for digito in n_str:
    suma += int(digito)
print(f"Suma dígitos: {suma}")
```

```python
n = int(input("¿Es primo?: "))
primo = n > 1
for i in range(2, int(n**0.5) + 1):
    if n % i == 0:
        primo = False
        break
print(f"Resultado: {primo}")
```

```python
n = int(input("Filas triángulo: "))
for i in range(1, n + 1):
    print("*" * i)
```

```python
b = int(input("Base: "))
e = int(input("Exponente: "))
res = 1
for _ in range(e):
    res *= b
print(f"Potencia: {res}")
```
