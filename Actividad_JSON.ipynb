##Taller Evaluador: Python y JSON


```python
animales = {"vacas": 10, "pollos": 15}
print(animales)
```

```python
import json

animales_json = json.dumps(animales)
print(animales_json)
```

```python
with open('granja.json', 'w') as f:
    json.dump(animales, f)
```

```python
frutas = ["manzana", "arandanos", "banano"]
frutas_json = json.dumps(frutas)
print(frutas_json)
```

```python
datos_red = '{"usuarios": 50, "likes": 120}'
diccionario_red = json.loads(datos_red)
print(diccionario_red["likes"])
```

```python
with open('granja.json', 'r') as f:
    lectura_granja = json.load(f)
    print(lectura_granja)
```

```python
vehiculo = {"marca": "Toyota", "modelo": 2024}
vehiculo_formateado = json.dumps(vehiculo, indent=4)
print(vehiculo_formateado)
```

```python
estudiante = {
    "nombre": "Amelia Pineda",
    "calificaciones": [95, 80, 100],
    "activo": True
}

with open('estudiante.json', 'w') as f:
    json.dump(estudiante, f)
```

```python
pais = input("Ingresa el nombre de un país: ")
capital = input("Ingresa su capital: ")

geografia = {"pais": pais, "capital": capital}

with open('geografia.json', 'w') as f:
    json.dump(geografia, f)
```

```python
with open('geografia.json', 'r') as f:
    datos_geo = json.load(f)

datos_geo["continente"] = "América"

with open('geografia.json', 'w') as f:
    json.dump(datos_geo, f, indent=4)
```
