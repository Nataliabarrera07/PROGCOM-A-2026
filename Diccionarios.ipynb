##DICCIONARIO
1. Tipo de dato como las listas **SI SON MUTABLES**
2. Se utiliza para almacenar datos en pares,
claves: valor, **La clave debe se ALFANUMERICA**


```
variable= {clave:valor}
```
Los diccionarios son colecciones ordenadas, modificables y **NO SE PERMITEN DUPLICADOS**


````python
estudiante= {"nombre":"Natalia",
             "ID":"U00091042",
             "Edad": 18,
             "Carrera": "Ingenieria industrial"}
print(estudiante)
````

````python
galletas= ["angelitos","diablitos","dinosaurios"]
galletas_precio=[20000,15000,25000]
galletas_cant=[3,5,10]
````

````python
def menu_galletas():
  for i in range(3):
    print(f"{galletas[i]}: {galletas_precio[i]},cantidad:{galletas_cant[i]}")
````

````python
menu_galletas()
````

````python
galleta= {"tipo de galleta":["angelitos","diablitos","dinosaurios"],
         "precio":[20000,15000,25000],
         "cantidad": [3,5,10]}
````

````python
def menu_galletas_dict():
  for i in range(3):
    print(f"{galletas["tipo de galleta"][i]}: {galleta["precio"][i]},cantidad:{galleta["cantidad"][i]}")
````

````python
menu_galletas_dict()
````
